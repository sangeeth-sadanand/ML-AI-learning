# Handling Missing Data

## Deletion method

```python
import seaborn as sns

# Load the 'tips' dataset
t_df = sns.load_dataset('titanic')
```

```python
t_df.head()
```

```python
t_df.describe(include='all')
```

```python
sns.heatmap(t_df.isnull(), cbar=False, cmap='viridis')
```

## Drop rows

```python
d_df = t_df.dropna(subset=["embarked"])
sns.heatmap(d_df.isnull(), cbar=False, cmap='viridis')
```

## Drop columns

```python
d_df = d_df.drop(columns=["deck"])
sns.heatmap(d_df.isnull(), cbar=False, cmap='viridis')
```

```python
df = d_df
df["age_mean"] = df['age'].fillna(df['age'].mean())
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
```

```python
df = d_df
df["age_mean"] = df['age'].fillna(df['age'].median())
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
```

```python
df = t_df
df["embarked_mean"] = df['embarked'].fillna(df['embarked'].mode()[0])
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
```
