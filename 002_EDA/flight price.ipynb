# Flight price  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/home/sans/Documents/002_ML-AI-learning/data/flight_price.xlsx")
df.head(5)
```

```python
# basic info about data
df.info()
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month']=df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year']=df['Date_of_Journey'].str.split('/').str[2].astype(int)
df.info()
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
df
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(" ")[0])
df['Arrival_Hour'] = df['Arrival_Time'].str.split(":").str[0].astype(int)
df['Arrival_Minute'] = df['Arrival_Time'].str.split(":").str[1].astype(int)
df.head(3)
```

```python
df.drop("Arrival_Time", inplace= True, axis=1)
```

```python
df['dep_Hour'] = df['Dep_Time'].str.split(":").str[0].astype(int)
df['dep_Minute'] = df['Dep_Time'].str.split(":").str[1].astype(int)
df.head(3)
```

```python
df.drop("Dep_Time", inplace= True, axis=1)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop': 0, '2 stops': 2, '1 stop': 1, '3 stops':3, np.nan: 1, '4 stops': 4})
```

```python
df.drop('Route', axis=1, inplace=True)
df.head(3)
```

```python
df['Duration_hour'] = df['Duration'].str.extract(r'(?P<Duration_hour>\d+)h').fillna(0)
df['Duration_minute'] = df['Duration'].str.extract(r'(?P<m>\d+)m').fillna(0)
```

```python
df.drop('Duration', axis=1, inplace=True)
df.head(3)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()

pd.concat([df, pd.DataFrame(encoder.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray(), columns=encoder.get_feature_names_out())],
          axis=1)
```
